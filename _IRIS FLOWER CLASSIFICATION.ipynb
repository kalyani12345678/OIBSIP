```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
iris=pd.read_csv("iris.csv")
```

```python
iris.describe()
```

```python
sns.pairplot(iris,hue="Name",markers=['o','s','D'])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
features = iris[["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]]
target = iris["Name"]

# Split data into training, validation, and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
```

```python
from sklearn.svm import SVC
svn=SVC()
svn.fit(X_train,y_train)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
predictions=svn.predict(X_test)
accuracy=accuracy_score(y_test,predictions)
conf_matrix = confusion_matrix(y_test,predictions)
classification_rep = classification_report(y_test,predictions)

# Display the results
print(f"Logistic Regression Model Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_rep)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
predictions=dtc.predict(X_test)
accuracy_dtc=accuracy_score(y_test,predictions)
print("accuracy:\n",'%.2f'% accuracy_dtc)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
predictions=lr.predict(X_test)
accuracy_lr=accuracy_score(y_test,predictions)
print("accuracy :\n",'%.2f'% accuracy_lr)
```
